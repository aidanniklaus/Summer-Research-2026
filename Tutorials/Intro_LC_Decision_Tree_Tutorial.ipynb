## Step 1: Choose what to download from MAST
In order to make any graph we need data. For the stars we will be looking at, TPFs hold this data; specifically, how much light (or flux) is emitted over time. Some stars in the TESS catalog already have TPFs readily available in MAST. Most, however, do not. This means we must instead take from MAST the raw, full frame image (FFI) of the star and make it into a TPF ourselves.

This may initially seem complicated, but don't worry! This tutorial covers both situations, as well as how to know which situation you find yourself in.
